```python
path='https://www.rle.mit.edu/boib/news/'
htmlfile=urllib.request.urlopen(path)
Soup = BeautifulSoup(htmlfile, "html.parser")
print(Soup)
```

```python
news=Soup.select('body>div#mainsite > div#interiorcontent > div#interiorleft > p ')
links=Soup.select('body>div#mainsite > div#interiorcontent > div#interiorleft > p > a')
images=Soup.select('body>div#mainsite > div#interiorcontent > div#interiorleft > p > img')
videos=Soup.select('body>div#mainsite > div#interiorcontent > div#interiorleft > p > iframe')
print(images)
print(videos)

i = 0
data = []
for new in news:
    time=re.findall(r'<p style="margin-bottom:0px;">(\d+.\d+.\d+)<br/>', str(new), re.S)
#    print(str(new))
    img=images[i].get('src')
    video = videos[i].get('src')
    if time == []:
        continue
    else:
        if img == None:
            img = ''
        if video == None:
            video=''
        
        info={
        'time' : time,
        'title' : links[i].get_text(),
        'tlink' : links[i].get('href'),
        'plink' : img,
        'vlink' : video
        }
        i = i + 1
        data.append(info)
print(data)

#for link in links:
 #   print(link.get('href'))
  #  print(link.get_text())
```
